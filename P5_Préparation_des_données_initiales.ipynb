#### Analyse de la disponibilité alimentaire mondiale

```python
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import statsmodels.api as sm
import scipy.stats as st
```

```python
Y = df_dataset_fao["Dispo. Alim. (Kcal/personne)"]
X = df_dataset_fao[["Evolution population (%)"]]
X = X.copy() # On modifiera X, on en crée donc une copie
X['intercept'] = 1.
result = sm.OLS(Y, X).fit() # OLS = Ordinary Least Square (Moindres Carrés Ordinaire)
a,b = result.params['Evolution population (%)'],result.params['intercept']

#Affichage
fig=plt.figure(figsize=(12,8))
ax = plt.axes()
ax = ax.set(xlabel='Evolution population (%)', ylabel='Dispo. Alim. (Kcal/personne)')
plt.scatter(df_dataset_fao["Evolution population (%)"], Y)
plt.plot(np.arange(df_dataset_fao["Evolution population (%)"].min(),df_dataset_fao["Evolution population (%)"].max()),
         [a*x+b for x in np.arange(df_dataset_fao["Evolution population (%)"].min(),df_dataset_fao["Evolution population (%)"].max())], 
         color='red')
plt.title('Analyse bivariée de la disponibilité alimentaire en fonction de l\'évolution de la population 2015 - 2017', 
  loc='center', 
  fontsize=12)
plt.show()
```

```python
# Calcul du coefficient de corrélation linéaire (Pearson)
r = st.pearsonr(df_dataset_fao["Evolution population (%)"],df_dataset_fao["Dispo. Alim. (Kcal/personne)"])[0]
print('r =', round(r,3), '\nr²=', round(r**2,3))
```

On remarque une **dépendance importante** de la disponibilité alimentaire en Kcal/personne avec l'évolution de la population 2015/2017. **Plus la population évolue, plus la disponibilité régresse**.

```python
# Loi Normale
x_min = df_dataset_fao["Dispo. Alim. (Kcal/personne)"].min()
x_max = df_dataset_fao["Dispo. Alim. (Kcal/personne)"].max()

mean = np.mean(df_dataset_fao["Dispo. Alim. (Kcal/personne)"]) 
std = np.std(df_dataset_fao["Dispo. Alim. (Kcal/personne)"])

x = np.linspace(x_min, x_max, 100)
y = st.norm.pdf(x,mean,std)

# Histogramme
fig=plt.figure(figsize=(12,8))
ax = plt.axes()
ax = ax.set(xlabel='Dispo. Alim. (Kcal/personne)', ylabel='Densité')
plt.hist(df_dataset_fao["Dispo. Alim. (Kcal/personne)"], density=True, color="green", alpha=0.75)
plt.plot(x,y, color='coral', label="Loi Normale")
plt.grid(True)
plt.title('Répartition de la disponibilité alimentaire 2017', 
  loc='center', 
  fontsize=12)
plt.legend()
fig.show()
```

```python
# Le test de Kolmogorov Smirnov
st.ks_2samp(df_dataset_fao["Dispo. Alim. (Kcal/personne)"],list(np.random.normal(np.mean(df_dataset_fao["Dispo. Alim. (Kcal/personne)"]), np.std(df_dataset_fao["Dispo. Alim. (Kcal/personne)"]), 1000)))
```

```python
# Test de Shapiro
st.shapiro(df_dataset_fao["Dispo. Alim. (Kcal/personne)"])
```

Grâce à cet histogramme et au test de Shapiro, on constate que **la répartition de la disponibilité alimentaire ne suit pas rééllement la loi normale**.  Le test de Kolmorgov Smirnov est moins explicite. Cette donnée pourra être importante pour les statistiques inférencielles sur le jeu de données.

```python
# Courbe de Lorenz sur la disponibilité alimentaire
dispo_alim = df_dataset_fao["Dispo. Alim. (Kcal/personne)"].values

n = len(dispo_alim) # Taille de l'échantillon
lorenz = np.cumsum(np.sort(dispo_alim)) / dispo_alim.sum()
lorenz = np.append([0],lorenz) # La courbe de Lorenz commence à 0

# Calcul de l'indice de Gini
AUC = (lorenz.sum() -lorenz[-1]/2 -lorenz[0]/2)/n # Surface sous la courbe de Lorenz. Le premier segment (lorenz[0]) est à moitié en dessous de 0, on le coupe donc en 2, on fait de même pour le dernier segment lorenz[-1] qui est à moitié au dessus de 1.
S = 0.5 - AUC # surface entre la première bissectrice et la courbe de Lorenz
gini = 2*S

fig=plt.figure(figsize=(12,8))
plt.axes().axis('equal')
xaxis = np.linspace(0-1/n,1+1/n,n+1) #Il y a un segment de taille n pour chaque individu, plus 1 segment supplémentaire d'ordonnée 0. Le premier segment commence à 0-1/n, et le dernier termine à 1+1/n.
plt.plot(xaxis,lorenz,drawstyle='steps-post')
plt.plot(np.arange(0,1.2,0.2),np.arange(0,1.2,0.2), color='green', dashes=[6, 2], alpha=0.3)

plt.text(0.5,0.5,('gini = '+ str(round(gini,3))),horizontalalignment='center',
     verticalalignment='center', fontsize=14)

plt.title('Courbe de Lorenz pour la répartition de la disponibilité alimentaire', 
  loc='center', 
  fontsize=12)
plt.show()
```

On remarque avec cette courbe de Lorenz et un indice de Gini à 0.091 que **la répartition de la disponibilité alimentaire est plutôt égalitaire**.
